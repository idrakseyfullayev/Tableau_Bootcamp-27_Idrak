## PDF yaratmaq

```python
from PIL import Image
from reportlab.lib.pagesizes import A4
from reportlab.lib.utils import ImageReader
from reportlab.pdfgen import canvas
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.pdfbase import pdfmetrics

# Fayl yolu vÉ™ Ã§Ä±xÄ±ÅŸ faylÄ±
base_path = r'C:\Users\dell\Desktop\Tableau\Tableau_Course\Pictures\Ders2_'
output_pdf = r'C:\Users\dell\Desktop\Tableau\Tableau_Course\Ders2_2.pdf'

# BaÅŸlÄ±qlar (plt.title hissÉ™sindÉ™n Ã§Ä±xarÄ±lÄ±b)
titles = [
    "Create Calculated Field - FIXED",
    "Create Calculated Field - FIXED + SUM",
    "Calculated-i Dimension bÃ¶lmÉ™sinÉ™ gÃ¶ndÉ™rmÉ™k",
    "Create Calculated Field - Fixed_Customer_ID_Sum > 60000 Customer_ID-lÉ™r",
    "Calculated-i Dimension bÃ¶lmÉ™sinÉ™ gÃ¶ndÉ™rmÉ™k (60000 > olanlarÄ±n ID-lÉ™ri)",
    "Create Calculated Field - COUNTD + IF (60000 > olanlarÄ±n ID-lÉ™rin sayÄ±)",
    "Area Chart",
    "",
    "",
    "",
    "Area Chart - HÉ™r il Ã¼Ã§Ã¼n Percentage fo Total",
    "Compute Using - Customer Segment seÃ§mÉ™k",
    "Result 1",
    "Result 2",
    "Result 3 (Customer Segment-i Label-dÉ™n Ã§Ä±xartmaq lazÄ±mdÄ±r)",
    "Result 3 (Bar-da Customer Segment-i Label-a qaytarmaq lazÄ±mdÄ±r)"
]

# Arial fontunu qeyd edirik
pdfmetrics.registerFont(TTFont('Arial', r'C:\Windows\Fonts\arial.ttf'))

# PDF yaradÄ±rÄ±q
c = canvas.Canvas(output_pdf, pagesize=A4)
width, height = A4

# 19â€“34 ÅŸÉ™killÉ™ri PDF-É™ É™lavÉ™ edirik
for i in range(19, 35):
    full_path = f'{base_path}{i}.jpg'
    img = Image.open(full_path)

    # ÅžÉ™kil nisbÉ™tini hesablayÄ±rÄ±q
    aspect = img.width / img.height
    available_height = height - 90
    available_width = width - 40
    new_width = available_width
    new_height = new_width / aspect

    if new_height > available_height:
        new_height = available_height
        new_width = new_height * aspect

    # BaÅŸlÄ±ÄŸÄ± É™lavÉ™ et
    title_index = i - 19
    if title_index < len(titles):
        title = titles[title_index]
    else:
        title = ""

    if title:
        c.setFont("Arial", 14)
        c.drawCentredString(width / 2, height - 40, title)
        top_y = height - 70
    else:
        top_y = height - 40

    # ÅžÉ™kilin koordinatlarÄ±nÄ± tÉ™yin et
    x = (width - new_width) / 2
    y = top_y - new_height

    # ÅžÉ™kli yerlÉ™ÅŸdir
    img_reader = ImageReader(img)
    c.drawImage(img_reader, x, y, width=new_width, height=new_height)

    c.showPage()

# PDF faylÄ± tamamla
c.save()
print("âœ… PDF yaradÄ±ldÄ±:", output_pdf)
```
