## PDF yaratmaq

```python
from PIL import Image
from reportlab.lib.pagesizes import A4
from reportlab.lib.utils import ImageReader
from reportlab.pdfgen import canvas
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.pdfbase import pdfmetrics

# Fayl yolları
base_path = r'C:\Users\dell\Desktop\Tableau\Tableau_Course\Pictures\Ders3_'
output_pdf = r'C:\Users\dell\Desktop\Tableau\Tableau_Course\Ders_3_3_Static Set yaratmaq, Dynamic Set yaratmaq.pdf'

# Başlıqlar (plt.title hissələrindən götürülüb)
titles = [
    "Static Set yaratmaq 2-ci üsul",
    "",
    "Set-i Color-ə atmaq",
    "Dynamic Set yaratmaq üçün Vizualı hazırlmaq",
    "Dynamic Set yaratmaq",
    "'Use all' etmək Vacibdir",
    "40,0000-dən yuxarı olanlar",
    "Set-i Color-ə atmaq",
    "Rəngi Color-dən də Dəyişmək olur",
    "Set-i birbaşa Filterə atmaq",
    "Dynamic Set - Top 3 (Yaratma qaydası eynidir)",
    "Set-i Color atmaq",
    "Rəngi Dəyişmək qaydası",
    "Result",
    "Set-i Filter-ə həm birbaşa, həm də Color-dən atmaq olar",
    "Result"
]

# Arial fontunu qeyd edirik
pdfmetrics.registerFont(TTFont('Arial', r'C:\Windows\Fonts\arial.ttf'))

# PDF yaradırıq
c = canvas.Canvas(output_pdf, pagesize=A4)
width, height = A4

# 28–43 şəkilləri PDF-ə əlavə edirik
for i in range(28, 44):
    full_path = f'{base_path}{i}.jpg'
    img = Image.open(full_path)

    # Şəkil nisbətini hesablayırıq
    aspect = img.width / img.height
    available_height = height - 90
    available_width = width - 40
    new_width = available_width
    new_height = new_width / aspect

    if new_height > available_height:
        new_height = available_height
        new_width = new_height * aspect

    # Başlıq
    title_index = i - 28
    if title_index < len(titles):
        title = titles[title_index]
    else:
        title = ""

    if title:
        c.setFont("Arial", 14)
        c.drawCentredString(width / 2, height - 40, title)
        top_y = height - 70
    else:
        top_y = height - 40

    # Şəkilin koordinatları
    x = (width - new_width) / 2
    y = top_y - new_height

    # Şəkli yerləşdir
    img_reader = ImageReader(img)
    c.drawImage(img_reader, x, y, width=new_width, height=new_height)

    c.showPage()

# PDF-i tamamla
c.save()
print("✅ PDF yaradıldı:", output_pdf)
```
